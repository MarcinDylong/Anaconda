```python
import pandas as pd
```

```python
fuel = pd.read_csv('./data/fuel.csv', low_memory=False, usecols=[ 'Vehicle ID','Year','Make','Model','Class','Fuel Type',
                                                                 'Combined MPG (FT1)'])
fuel.head()
```

```python
fuel['Combined KPL'] = round(fuel['Combined MPG (FT1)'] * 1.609/3.78,2)
```

```python
fuel.head()
```

```python
fuel.insert(loc=5, column='liters per km', value = round(100 / fuel["Combined KPL"],2))
fuel.head()
```

```python
fuel.drop(labels='Combined MPG (FT1)', axis=1, inplace=True)
```

```python
del fuel['Combined KPL']
```
