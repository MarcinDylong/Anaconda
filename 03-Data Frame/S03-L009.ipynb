```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('./data/Fortune_500_2017.csv', usecols=['Rank','Title','Employees','Profits','Assets'],
                     index_col='Rank')
fortune.head()
```

```python
fortune['RankByEmp']=fortune['Employees'].rank(ascending=True)
fortune.sort_values(by='RankByEmp').head()
```

```python
fortune['RankByProf']=fortune['Profits'].rank(ascending=False)
fortune.sort_values(by='RankByProf').head()
```

```python
fortune.nlargest(3,'Assets').head()
```

```python
fortune.nsmallest(3,'Assets').head()
```
