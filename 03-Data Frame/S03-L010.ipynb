```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('./data/Fortune_500_2017.csv', usecols=['Rank','Title','Employees','Profits','Assets'], index_col='Rank')
fortune.head()
```

```python
fortune['RankByEmp'] = fortune['Employees'].rank(ascending=False)
fortune['RankByProf'] = fortune['Profits'].rank(ascending=False)
fortune.head()
```

```python
isEmployeesRankFirst10 = fortune['RankByEmp'] <= 10
isProfitRankFirst10 = fortune['RankByProf'] <= 10
```

```python
fortune[isEmployeesRankFirst10 & isProfitRankFirst10]
```

```python
isEmployeesRankMore400 = fortune['RankByEmp'] >= 400
```

```python
fortune[isEmployeesRankMore400 & isProfitRankFirst10]
```
