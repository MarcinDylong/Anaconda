```python
import pandas as pd
import numpy as np
```

```python
animals = pd.read_csv('./data/sleep_time.csv', index_col='ID').fillna(value=0)
animals.head()
```

```python
herbi = animals.where(animals['vore']=='herbi')
print(herbi.size)
print(herbi.dropna().size)
```

```python
omni = animals.where(animals['vore']=='omni')
print(omni.size)
print(omni.dropna().size)
```

```python
herbi = animals[animals['vore']=='herbi']
omni = animals[animals['vore']=='omni']
carni = animals[animals['vore']=='carni']
print('Herbi mean total sleep:',herbi['sleep_total'].mean())
print('Omni mean total sleep:',omni['sleep_total'].mean())
print('Carni mean total sleep:',carni['sleep_total'].mean())
```
