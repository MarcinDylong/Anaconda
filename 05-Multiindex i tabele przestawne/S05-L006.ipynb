```python
import pandas as pd
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./data/marathon_results_2016.csv', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],
                index_col='Bib')
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
df.info()
```

```python
df.pivot_table(index='Age', columns='M/F', values='TotalSeconds').head()
```

```python
df.pivot_table(index='Age', columns='M/F', values=['TotalSeconds','HalfSeconds']).head()
```

```python
df.pivot_table(index='Age', columns='M/F', values=['TotalSeconds','HalfSeconds']).sort_index(axis=1, level=1).head()
```

```python
df.pivot_table(index='Age', columns='M/F', values='Country', aggfunc='count').fillna(0)
```

```python
df.pivot_table(index='Age', values='Country', aggfunc='count').fillna(0)
```

```python
df.pivot_table(index='Country', values='Age', aggfunc='count').fillna(0)
```

```python
df.pivot_table(index='Country', values='Age', aggfunc='count').fillna(0).sort_values(by='Age', ascending=False)
```

```python
df.pivot_table(index=['Country','State','City'], values='TotalSeconds', aggfunc='mean').sort_values(by='TotalSeconds', ascending=False)
```
