```python
import pandas as pd
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./data/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
df = df.head(1000)
group_city = df.groupby("City")
group_city.size()
```

```python
for city in group_city:
    print(city[0])
```

```python
for city, data in group_city:
    print(city," - ",len(data))
```

```python
for city, data in group_city:
    print(city," - ", data.nsmallest(1,'TotalSeconds'))
```

```python
the_best_per_city = pd.DataFrame()
```

```python
for city, data in group_city:
    the_best_per_city = the_best_per_city.append(data.nsmallest(1,'TotalSeconds'))
    
the_best_per_city
```
