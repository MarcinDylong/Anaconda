```python
import pandas as pd
import numpy as np
```

```python
customers_names = pd.read_csv('./data/northwind-mongo-master/customers.csv', usecols=['CustomerID','CompanyName'])
customers_cities = pd.read_csv('./data/northwind-mongo-master/customers.csv', usecols=['CustomerID','City','Country'])
```

```python
print(len(customers_names))
print(len(customers_cities))
```

```python
customers_names.set_index(keys = 'CustomerID', inplace=True)
customers_cities.set_index(keys = 'CustomerID', inplace=True)
```

```python
customers_names.join(other=customers_cities)
```
