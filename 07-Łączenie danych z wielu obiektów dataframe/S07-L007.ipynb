```python
import pandas as pd
import numpy as np
```

```python
suppliers = pd.read_csv('./data/northwind-mongo-master/suppliers.csv', usecols=['SupplierID','CompanyName','City'])
suppliers.head()
```

```python
products = pd.read_csv('./data/northwind-mongo-master/products.csv', usecols=['ProductID','ProductName','SupplierID'])
products.head()
```

```python
sup_prod = suppliers.merge(products)
sup_prod.head()
```

```python
suppliers.set_index(keys='SupplierID', inplace=True)
sup_prod = suppliers.merge(products, left_index = True, right_on='SupplierID')
sup_prod.head()
```

```python
products.set_index(keys='SupplierID', inplace=True)
sup_prod = suppliers.merge(right=products, left_index=True, right_index=True)
sup_prod.head()
```

```python
suppliers.reset_index(inplace=True)
sup_prod = suppliers.merge(right=products, left_on='SupplierID', right_index=True)
sup_prod.head()
```

```python
products.reset_index(inplace=True)
suppliers.rename({'SupplierID':'suppliers'}, axis='columns', inplace=True)
sup_prod = suppliers.merge(right=products, left_on='suppliers', right_on='SupplierID')
sup_prod.head()
```
