```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv('./data/results.csv')
data
```

```python
pl_data = data.query("home_team=='Poland' or away_team =='Poland'")
pl_data.reset_index(inplace=True)
pl_data.drop('index', axis=1, inplace=True)
pl_data.info(memory_usage='deep')
```

```python
def opponent(row):
    if row['home_team'] == 'Poland':
        opp = row['away_team']
    else:
        opp =  row['home_team']
    return opp

pl_data['opponent'] = pl_data.apply(opponent, axis=1)
pl_data['scored'] = pl_data.apply(lambda x: x['home_score'] if x['home_team'] == 'Poland' else x['away_score'], axis=1)
pl_data['conceded'] = pl_data.apply(lambda x: x['away_score'] if x['home_team'] == 'Poland' else x['home_score'], axis=1)
pl_data['is_home'] = pl_data.apply(lambda x: True if (x['home_team']=='Poland' and x['neutral'] == False) else False, axis=1)
pl_data.head()
```

```python
to_drop = ['home_team','away_team','home_score','away_score']
pl_data.drop(to_drop,inplace=True, axis=1)
pl_data = pl_data[['date','tournament','opponent','scored','conceded','is_home','neutral','city','country']]
pl_data.head()
```

```python
pl_data.info(memory_usage='deep')
```

```python
pl_data['tournament'] = pl_data['tournament'].astype('category')
pl_data['opponent'] = pl_data['opponent'].astype('category')
pl_data['city'] = pl_data['city'].astype('category')
pl_data['country'] = pl_data['country'].astype('category')
pl_data.info(memory_usage="deep")
```

```python
pl_data.to_csv('./data/results_pl.csv', index=False)
```
